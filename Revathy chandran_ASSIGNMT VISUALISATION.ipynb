```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#load data set
data=pd.read_csv(r"C:\Users\user\Downloads\data (1).csv")
```

```python
#to know shape of data
data.shape
```

```python
#check for null values
data.isna().sum()
```

```python
#replace null values in calories with median
data['Calories'].fillna(data['Calories'].median(),inplace=True)
```

```python
data
```

```python
#check null values
data.isna().sum()
```

```python
#corrected the date format
data['Date']=pd.to_datetime(data['Date'])
```

```python
data
```

```python
#drop row with null values in date
data.dropna(subset=['Date'],inplace=True)
```

```python
#corect the typing error in loc 7
data.loc[7,'Duration']=45
```

```python
data
```

```python
#check whether missing values are cleared
data.isna().sum()
```

```python
#check for duplicate values
data.duplicated()
```

```python
#drop duplicate values
data.drop_duplicates(inplace=True)
```

```python
#check correlation
data.corr()
```
