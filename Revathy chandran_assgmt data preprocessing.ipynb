Make â€˜PassengerIdâ€™ as the index column

```python
data.set_index("PassengerId", inplace = True)
```

```python
data.head(5)
```

```python
#float values stored to num
num = data[[ 'Age','Fare']]
```

```python
#categorical values stored to obj
obj= data[[ 'Name', 'Sex', 'Ticket','Cabin', 'Embarked']]

```

```python
#filling numerical values with median and categorical values with mode
```

```python
for i in num:
    data[i].fillna(data[i].median(),inplace=True)
```

```python
for i in obj:
    data[i].fillna(data[i].value_counts().index[0],inplace=True)
```

```python
#afer filing null values check for null values
data.isna().sum()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder
lab_enc=LabelEncoder()
data['Name']=lab_enc.fit_transform(data['Name'])
data['Sex']=lab_enc.fit_transform(data['Sex'])
data['Ticket']=lab_enc.fit_transform(data['Ticket'])
data['Cabin']=lab_enc.fit_transform(data['Cabin'])
data['Embarked']=lab_enc.fit_transform(data['Embarked'])
```
